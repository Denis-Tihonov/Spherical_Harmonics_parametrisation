```python
accel_ssa = SSA(x_acc, 500)

x_acc_clear = accel_ssa.reconstruct(slice(0,5))

def HankelMatrix(X, L):  
    N = X.shape[0]
    return scipy.linalg.hankel(X[ : N - L + 1], X[N - L : N])
```

```python
fig = go.Figure()

fig.add_scatter(y = x_acc,
                mode='lines',
                name='Full')

fig.add_scatter(y = x_acc_clear,
                mode='lines',
                name='0-5')

fig.show()
```
